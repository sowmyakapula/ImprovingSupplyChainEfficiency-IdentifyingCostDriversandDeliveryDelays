# Exploratory Data Analysis: Supply Chain Optimization

## Objective
The goal of this analysis is to identify key factors driving higher costs and delivery delays in the supply chain using exploratory data analysis techniques.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("cleaned_dataset.csv")
df.head()
```

## Univariate Analysis

```python
df.hist(figsize=(12,10))
plt.show()
```

Most variables show [normal/skewed] distributions. Costs and shipping costs appear right-skewed, indicating the presence of higher-value outliers.

## Correlation Analysis

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

## Bivariate Analysis

### Cost vs Shipping Time

```python
sns.scatterplot(x='shipping_times', y='costs', data=df)
plt.show()
```

### Product Type vs Cost

```python
sns.boxplot(x='product_type', y='costs', data=df)
plt.show()
```

### Location vs Delay

```python
sns.boxplot(x='location', y='shipping_times', data=df)
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df[['costs','shipping_costs','shipping_times','defect_rates']])
plt.show()
```

## Feature Insights

## Key Insights

1. Higher shipping costs are associated with increased total costs.
2. Certain product types consistently show higher costs.
3. Some locations experience longer delivery times.
4. Transportation modes impact both cost and delivery efficiency.

## Feature Importance

Based on the analysis, the following variables are likely to be important for modeling:

- shipping_costs
- shipping_times
- product_type
- location
- transportation_modes
